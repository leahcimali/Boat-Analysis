```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
plt.style.use('seaborn-white')
```

```python
df = pd.read_csv('boat_data.csv')
sort = df.sort_values('Number of views last 7 days',ascending = 0)
sort.describe()
```

```python
sort.head()
```

```python
sort['Material'].value_counts()
```

```python
sort['Currency']=sort['Price'].apply(lambda x: x.split(' ')[0])
sort['Currency'].value_counts()
```

```python
sort['Type'].head(1000).value_counts()
```

```python
sort['Boat Type'].value_counts().head(10)
```

```python
def conversion_eur(L):
    if L[0] == 'CHF':
        return int(L[1])*0.96
    elif L[0] == 'DKK':
        return int(L[1])*0.13
    elif L[0] == 'Â£':
        return int(L[1])*1.19
    else:
        return int(L[1])
```

```python
sort['Price Euro (1000K)'] = sort['Price'].apply(lambda x: conversion_eur(x.split(' ')))*0.001
sort[['Price Euro (1000K)','Price','Currency']].head(100)
```

```python
sort['Used or New']=sort['Type'].apply(lambda x: str(x).split(',')[0])
sort['Motor Type']=sort['Type'].apply(lambda x: str(x).split(',')[1] if len(str(x).split(','))>1 else None)
sort['Location']=sort['Location'].apply(lambda x: str(x).split(' ')[0])
sort['Location'].unique()
```

```python
sort["Location"]= sort["Location"].replace(['Thun','Lake','ZÃ¼richse', 'Avenches'],'Switzerland')
sort["Location"]= sort["Location"].replace(['Neustadt','TravemÃ¼nde','Steinwiesen','Rheinfelden'],'Germany')
```

```python
sort.loc[sort["Used or New"]=='new boat from stock', 'Used or New']='New'
sort.loc[sort["Used or New"]=='new boat on order', 'Used or New']='New'
sort.loc[sort["Used or New"]=='Used boat', 'Used or New']='Used'
sort["Used or New"] = sort["Used or New"].replace(['Display Model','Diesel', 'Unleaded', 'nan', 'Electric'] , 'Other')
sort = sort.drop(['Price','Type','Currency'],axis =1)
```

```python
sort["Motor Type"].value_counts()
```

```python
sort['Decade']=sort['Year Built'].apply(lambda x: int(x)//10*10)
sort['Decade'].value_counts()
```

```python
goal_mean = 151.819752 * 1.75
more = sort[sort['Number of views last 7 days']> goal_mean]
more.head(100)
```

```python
more[['Boat Type','Number of views last 7 days']].groupby('Boat Type').sum().sort_values('Number of views last 7 days',ascending = False)
```
